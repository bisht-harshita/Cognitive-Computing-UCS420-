```python
# Q1
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Part I: Randomized Sales Data Generation
# Initialize Random Seed using Roll Number
np.random.seed(102317053)

# Generate Sales Data for 12 months and 4 categories
sales_data = np.random.randint(1000, 5000, size=(12, 4))

# Convert to DataFrame
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
categories = ['Electronics', 'Clothing', 'Home & Kitchen', 'Sports']
sales_df = pd.DataFrame(sales_data, columns=categories, index=months)

print("First 5 rows of Sales Data:")
print(sales_df.head())

# Summary Statistics
print("\nSummary Statistics:")
print(sales_df.describe())

# Part II: Data Manipulation and Analysis
# Total Sales per Category
category_sales = sales_df.sum(axis=0)
print("\nTotal Sales per Category:")
print(category_sales)

# Total Sales per Month
monthly_sales = sales_df.sum(axis=1)
sales_df['Total Sales'] = monthly_sales
print("\nTotal Sales per Month:")
print(monthly_sales)

# Average Sales Growth Calculation
sales_df['Growth Rate'] = sales_df['Total Sales'].pct_change() * 100

# Apply Discount Based on Roll Number
roll_number = 102317053
if roll_number % 2 == 0:
    sales_df['Electronics'] *= 0.9
else:
    sales_df['Clothing'] *= 0.85

print("\nSales Data After Discount:")
print(sales_df)

# Part III: Visualization
plt.figure(figsize=(10, 6))
sns.lineplot(data=sales_df[categories])
plt.title('Monthly Sales Trends by Category')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.legend(categories)
plt.grid(True)
plt.show()

# Box Plot for Sales Distribution
plt.figure(figsize=(8, 5))
sns.boxplot(data=sales_df[categories])
plt.title('Sales Distribution by Category')
plt.show()
```

```python
# Q2
array = np.array([[1, -2, 3], [-4, 5, -6]])
print("\nElement-wise Absolute Value:")
print(np.abs(array))

print("\nPercentiles:")
print("Flattened Array - 25th, 50th, 75th:", np.percentile(array, [25, 50, 75]))
print("Column-wise Percentiles:", np.percentile(array, [25, 50, 75], axis=0))
print("Row-wise Percentiles:", np.percentile(array, [25, 50, 75], axis=1))

print("\nMean, Median, and Std Deviation:")
print("Flattened: Mean =", np.mean(array), "Median =", np.median(array), "Std =", np.std(array))
print("Column-wise:", np.mean(array, axis=0), np.median(array, axis=0), np.std(array, axis=0))
print("Row-wise:", np.mean(array, axis=1), np.median(array, axis=1), np.std(array, axis=1))
```
