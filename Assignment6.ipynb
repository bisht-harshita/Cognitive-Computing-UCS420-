```python
import numpy as np
import matplotlib.pyplot as plt

# Ask the user to enter a value for M
M = float(input("Enter the value of M: "))

# Generate x values from -10 to 10
x = np.linspace(-10, 10, 100)

# Compute y values
y1 = M * x**2  # y = M * x^2
y2 = M * np.sin(x)  # y = M * sin(x)

# Create the plot
plt.figure(figsize=(10, 5))

# Plot y = M * x^2
plt.plot(x, y1, 'b--', label='y=M*x^2')  # Blue dashed line

# Plot y = M * sin(x)
plt.plot(x, y2, 'r-', label='y=M*sin(x)')  # Red solid line

# Labels and title
plt.xlabel("x values")
plt.ylabel("y values")
plt.title(f"Graphs of y = {M}x and y = {M}sin(x)")

# Add legend and grid
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)

# Show the plot
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Step 1: Create the dataset
data = {
    "Subject": ["Math", "Science", "English", "History", "Computer Science"],
    "Score": [85, 92, 78, 88, 95]
}

# Step 2: Convert to Pandas DataFrame
df = pd.DataFrame(data)

# Step 3: Set Seaborn style
sns.set_style("whitegrid")

# Step 4: Create the bar plot
plt.figure(figsize=(8, 5))
colors = ["blue", "green", "red", "purple", "orange"]
ax = sns.barplot(x="Subject", y="Score", hue="Subject", data=df, palette=colors, legend=False)


# Step 5: Add annotations (scores on top of bars)
for p in ax.patches:
    ax.annotate(f'{p.get_height()}', 
                (p.get_x() + p.get_width() / 2, p.get_height()), 
                ha='center', va='bottom', fontsize=12, fontweight='bold', color='black')

# Step 6: Add title, labels, and grid
plt.title("Student Scores in Different Subjects", fontsize=14, fontweight='bold')
plt.xlabel("Subjects", fontsize=12)
plt.ylabel("Scores", fontsize=12)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Step 7: Show the plot
plt.show()
```
